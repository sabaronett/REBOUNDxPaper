# Figure 1

> Post-MS mass (top) and radius (bottom) evolution of a Sun-like star, with various Reimers scaling factors (Eq. 3), from `MESA`.  At about two-thirds in age from each TRGB to TAGB, tiny drops in mass and transient bumps in radius correspond to thermal pulses along the AGB resolved by `MESA`.  For reference, $1~\textrm{au} \approx 215~R_{\odot}$.

## Load

We first load in the extracted data, exported from MESA log files, as done in [mesa2txt.ipynb](https://github.com/sabaronett/REBOUNDxPaper/blob/master/mesa2txt.ipynb).

```python
import numpy as np

data = np.loadtxt('mesa/data/eta_0.2/m.txt') # return (N, 2) array
mtimes02 = data[:, 0]                        # return only 1st col
masses02 = data[:, 1]                        # return only 2nd col
data = np.loadtxt('mesa/data/eta_0.5/m.txt')
mtimes05 = data[:, 0]
masses05 = data[:, 1]
data = np.loadtxt('mesa/data/eta_0.8/m.txt')
mtimes08 = data[:, 0]
masses08 = data[:, 1]
data = np.loadtxt('mesa/data/eta_0.2/r.txt')
rtimes02 = data[:, 0]
Rsuns02 = data[:, 1]                         # data in Rsun units
data = np.loadtxt('mesa/data/eta_0.5/r.txt')
rtimes05 = data[:, 0]
Rsuns05 = data[:, 1]
data = np.loadtxt('mesa/data/eta_0.8/r.txt')
rtimes08 = data[:, 0]
Rsuns08 = data[:, 1]
```

## Plot

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(6, 5))
fig.subplots_adjust(hspace=0)

ax1.set_ylabel("$M(t)$ / $M_{\odot}$", fontsize='large')
ax1.yaxis.set_minor_locator(mticker.AutoMinorLocator())
ax1.plot(mtimes02/1e9, masses02, label=r'$\eta=0.2$')
ax1.plot(mtimes05/1e9, masses05, label=r'$\eta=0.5$')
ax1.plot(mtimes08/1e9, masses08, label=r'$\eta=0.8$')
ax1.grid()
ax1.legend(loc='upper right')

ax2.set_xlabel('Age / Gyr', fontsize='large')
ax2.xaxis.set_minor_locator(mticker.AutoMinorLocator())
ax2.set_ylabel('$R(t)$ / $R_{\odot}$', fontsize='large')
ax2.yaxis.set_minor_locator(mticker.AutoMinorLocator())
ax2.plot(rtimes02/1e9, Rsuns02)
ax2.plot(rtimes05/1e9, Rsuns05)
ax2.plot(rtimes08/1e9, Rsuns08)
ax2.grid()

plt.savefig('img/fig1.eps', bbox_inches='tight', pad_inches=0.01)
plt.savefig('img/fig1.pdf', bbox_inches='tight', pad_inches=0.01)
```
